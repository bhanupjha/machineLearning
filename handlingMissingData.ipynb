```python
import pandas as pd
data1 = pd.read_csv("Titanic-Dataset.csv")

data1.dropna(inplace=True)



```

```python
data1.head(20)
```

```python
# technique 2
# imputation
data1['Age'].fillna(data1['Age'].mean(), inplace=True)
data1.head(20)
```

```python
# filling median
data1['Age'].fillna(data1['Age'].median(), inplace=True)
data1.head(20)
```

```python
# filling mode
data1['Age'].fillna(data1['Age'].mode(), inplace=True)
data1.head(20)
```

```python
# technique 3
# Linear interpolation
data1.interpolate(method='linear', inplace=True)
data1.head(10)
```

```python
# technique 4

# forward  fill

data1.fillna(method='ffill', inplace=True)
data1.head(10)
```

```python
# backward fill

data1.fillna(method='bfill', inplace=True)
data1.head(10)
```
