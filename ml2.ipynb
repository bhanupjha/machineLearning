```python
import  pandas as pd
data = pd.read_csv("housing.csv")
data.head(60)
```

```python
data
```

```python
data.isnull().any()
```

```python
# data.dropna(inplace=True)
# data.head(10)

# replace null value by mean of that column
data['total_bedrooms'].fillna(data['total_bedrooms'].mean(), inplace=True)
data.head(20)
```

```python
data['ocean_proximity'].value_counts
```

```python
# for removing catageroical data
data_encoded = pd.get_dummies(data,columns=['ocean_proximity'],prefix = 'Is')
print(data_encoded)
```

```python
# splitting into training and testing
from sklearn.model_selection import train_test_split

x = data_encoded.drop(columns = 'median_house_value', inplace = False)
y = data_encoded['median_house_value']

x_train , x_test, y_train , y_test = train_test_split(x,y,test_size = 0.2, random_state= 0)
```

```python
print(x_train.shape)
print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train , y_train)
```

```python
# print the intercept and coefficient
print(model.intercept_)
print(model.coef_)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```
